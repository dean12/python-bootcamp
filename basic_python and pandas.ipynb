```python
# ### Generate 25 unique random numbers between 1 and 100. 
# Find min, max, mean and median.

import numpy as np

# Lets use a list 
rand_numbers = []  #can also use list()

while len(rand_numbers) < 25:
        #Generate a random number between 1 and 100
        rand_no = rnd.randint(1, 100)
        #Add number to list if the number is not already in the list
        if rand_no not in rand_numbers:
           rand_numbers.append(rand_no)
            
            
print('Number list is ', rand_numbers)
print('Count ', len(rand_numbers))

print('Max is ', max(rand_numbers))
print('Min is ', min(rand_numbers))
print('Average is ', np.mean(rand_numbers))
print('Median is ', np.median(rand_numbers))
```

```python
rand_numbers = set()

while len(rand_numbers) < 25:
        #Add number to set
        rand_no = rnd.randint(1, 100)
        rand_numbers.add(rand_no)
            
            
print('Number list is ', rand_numbers)
print('Count ', len(rand_numbers))

print('Max is ', max(rand_numbers))
print('Min is ', min(rand_numbers))
print('Average is ', np.mean(list(rand_numbers)))
print('Average is ', np.median(list(rand_numbers)))
```

```python
# Predict a number that comes up when you roll a dice three times. 
# Print 'You win' if your number comes one time
# Print 'Two times a winner' if the number comes two times
# Print 'Jackpot' if it comes three times


# Get input from user
your_bet = 3

outcome = []
for x in range(0,3):
    dice_number = rnd.randint(1, 6)
    outcome.append(dice_number)

bet_count = outcome.count(int(your_bet))

print(bet_count)

if bet_count == 1:
    print('You win')
elif bet_count == 2:
    print('Two times a winner')
elif bet_count == 3:
    print('Jackpot!')
else:
    print('You lose')    
```
